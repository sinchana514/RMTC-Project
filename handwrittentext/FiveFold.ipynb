```python
import warnings
warnings.filterwarnings("ignore")

import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix
import seaborn as sns
%matplotlib inline

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python

# Reshape the input data
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)
```

```python
# Define the input and output dimensions
input_dims = X_train.shape[1]
output_dims = 10

# Define the number of epochs and batch size
num_epochs = 20
batch_size = 128

# Define the number of folds for cross-validation
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits)

# Initialize lists to store training/validation scores
train_acc_list = []
val_acc_list = []
train_loss_list = []
val_loss_list = []
```

```python
# Perform 5-fold cross-validation
fold = 0
for train_index, val_index in skf.split(X_train, y_train):
    fold += 1
    print(f'Fold {fold}/{n_splits}')

    X_train_fold, X_val_fold = X_train[train_index], X_train[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    # Define the model architecture
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(512, activation='sigmoid', input_shape=(input_dims,)),
        tf.keras.layers.Dense(128, activation='sigmoid'),
        tf.keras.layers.Dense(output_dims, activation='softmax')
    ])

    # Compile the model
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    # Convert labels to one-hot encoding
    y_train_fold_one_hot = tf.keras.utils.to_categorical(y_train_fold, num_classes=output_dims)
    y_val_fold_one_hot = tf.keras.utils.to_categorical(y_val_fold, num_classes=output_dims)

    # Train the model
    history = model.fit(X_train_fold, y_train_fold_one_hot, validation_data=(X_val_fold, y_val_fold_one_hot), epochs=num_epochs, batch_size=batch_size, verbose=1)

    # Evaluate the model on test data
    score = model.evaluate(X_test, tf.keras.utils.to_categorical(y_test, num_classes=output_dims), verbose=0)
    print(f'Test score for Fold {fold}:', score[0])
    print(f'Test accuracy for Fold {fold}:', score[1])

    # Predict the labels for the test set
    y_pred = np.argmax(model.predict(X_test), axis=1)
    # Generate confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Confusion Matrix - Fold {fold}')
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.show()

     # Append training and validation metrics to lists
    train_acc_list.append(history.history['accuracy'])
    val_acc_list.append(history.history['val_accuracy'])
    train_loss_list.append(history.history['loss'])
    val_loss_list.append(history.history['val_loss'])

# Calculate average training and validation metrics across all folds
avg_train_acc = np.mean(train_acc_list, axis=0)
avg_val_acc = np.mean(val_acc_list, axis=0)
avg_train_loss = np.mean(train_loss_list, axis=0)
avg_val_loss = np.mean(val_loss_list, axis=0)
```

```python
# Plot average training and validation accuracy
plt.plot(avg_train_acc)
plt.plot(avg_val_acc)
plt.title('Average Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'])
plt.show()

```

```python
# Plot average training and validation loss
plt.plot(avg_train_loss)
plt.plot(avg_val_loss)
plt.title('Average Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'])
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")

import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix
import seaborn as sns
%matplotlib inline

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Reshape the input data
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)

# Define the input and output dimensions
input_dims = X_train.shape[1]
output_dims = 10

# Define the number of epochs and batch size
num_epochs = 20
batch_size = 128

# Define the number of folds for cross-validation
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits)

# Initialize lists to store training/validation scores
train_acc_list = []
val_acc_list = []
train_loss_list = []
val_loss_list = []

# Initialize variables to store accumulated predictions and true labels
accumulated_y_true = []
accumulated_y_pred = []

# Perform 5-fold cross-validation
fold = 0
for train_index, val_index in skf.split(X_train, y_train):
    fold += 1
    print(f'Fold {fold}/{n_splits}')

    X_train_fold, X_val_fold = X_train[train_index], X_train[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    # Define the model architecture
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(512, activation='sigmoid', input_shape=(input_dims,)),
        tf.keras.layers.Dense(128, activation='sigmoid'),
        tf.keras.layers.Dense(output_dims, activation='softmax')
    ])

    # Compile the model
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    # Convert labels to one-hot encoding
    y_train_fold_one_hot = tf.keras.utils.to_categorical(y_train_fold, num_classes=output_dims)
    y_val_fold_one_hot = tf.keras.utils.to_categorical(y_val_fold, num_classes=output_dims)

    # Train the model
    history = model.fit(X_train_fold, y_train_fold_one_hot, validation_data=(X_val_fold, y_val_fold_one_hot), epochs=num_epochs, batch_size=batch_size, verbose=1)

    # Evaluate the model on test data
    score = model.evaluate(X_test, tf.keras.utils.to_categorical(y_test, num_classes=output_dims), verbose=0)
    print(f'Test score for Fold {fold}:', score[0])
    print(f'Test accuracy for Fold {fold}:', score[1])

    # Predict the labels for the test set
    y_pred = np.argmax(model.predict(X_test), axis=1)
    accumulated_y_true.extend(y_test)
    accumulated_y_pred.extend(y_pred)

    # Append training and validation metrics to lists
    train_acc_list.append(history.history['accuracy'])
    val_acc_list.append(history.history['val_accuracy'])
    train_loss_list.append(history.history['loss'])
    val_loss_list.append(history.history['val_loss'])

# Generate confusion matrix for all folds combined
cm_all_folds = confusion_matrix(accumulated_y_true, accumulated_y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_all_folds, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - All Folds')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()

# Calculate average training and validation metrics across all folds
avg_train_acc = np.mean(train_acc_list, axis=0)
avg_val_acc = np.mean(val_acc_list, axis=0)
avg_train_loss = np.mean(train_loss_list, axis=0)
avg_val_loss = np.mean(val_loss_list, axis=0)


```
